```python
from bs4 import BeautifulSoup

import requests

import csv

url = requests.get('https://pokemondb.net/pokedex/all')

soup = BeautifulSoup(url.content, 'html.parser')

csv_file = open('Pokédex.csv', 'w', encoding='utf-8')
csv_writer = csv.writer(csv_file)
csv_writer.writerow(['#', 'Name', 'Type', 'Base Total'])

for x in soup.find_all('tr'):
    
    no = x.find('span', class_='infocard-cell-data')
    if no!=None:
        nos = no.text
        print(nos)
    
    name = x.find('a', class_='ent-name')
    if name!=None:
        names = name.text
        print(names)
    
    type = x.find(class_='cell-icon')
    if type!=None:
        types = type.text
        print(types)
    
    base_total = x.find(class_='cell-total')
    if base_total!=None:
        bases = base_total.text
        print(bases)
    csv_writer.writerow([nos, names, types, bases])

csv_file.close()
```
